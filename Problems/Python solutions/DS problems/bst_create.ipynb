```python
class Node:

  def __init__(self, value):
    self.val = value
    self.left_child = None
    self.right_child = None


  def insert(self, data):
    if(data == self.val):
      return False

    elif(data < self.value):
      if(self.left_child):
        self.left_child.val = self.insert(data)
        return True
      else:
        self.left_child = Node(data)

    else:
      if(self.right_child):
        self.right_child.val = self.insert(data)
        return True
      else:
        self.right_child = Node(data)


  def find(self, data):
    if(self.val == data):
      return True

    elif(data < self.val):
      if(self.left_child):
        self.left_child.find(data)
      else:
        return False

    else:
      if(self.right_child):
        self.right_child.find(data)
      else:
        return False


  def preorder(self):
    print(self.val)
    self.left_child.preorder()
    self.right_child.preorder()


  def postorder(self):
    self.right_child.postorder()
    self.right_child.postorder()
    print(self)


  def inorder(self):
    self.left_child.inorder()
    print(self.val)
    self.right_child.inorder()

class Tree:

  def __init__():
    self.root = None

  def insert(self, data):

    if(self.root):
      return self.root.insert(data)
    else:
      return self.root = Node(data)

  def find(self, data):
    if(self.root):
      return self.root.find(data)
    else:
      return False

  def preorder():
    if(self.root):
      self.root.preorder()

  def postorder():
    if(self.root):
      self.root.postorder()

  def inorder():
    if(self.root):
      self.root.inorder()
      
```
