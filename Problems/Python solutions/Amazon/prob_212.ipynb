```python
class Solution:
    def findWords(self, board, words):
        final_words_list = []
        for word in words:
            if(self.findWord(board, word) == True):
                final_words_list.append(word)
        return final_words_list
                
    def findWord(self, board, word):
        firstLetter = word[0]
        for r in range(len(board)):
            for c in range(len(board[0])):
                if(firstLetter == board[r][c]):
                    print("\nChecking Row {row} and Column {col}".format(row = r, col = c))
#                     print("Word to be checked..." + word)
                    visited = set([])
#                     print("Visited set..." + str(visited))
                    if(self.checkLetter(board, r, c, word, firstLetter, 1, visited)[0] == True):
                        return True
        return False
                            
    def checkLetter(self, board, r, c, checkWord, formedWord, letterIndex, visited):
        if(formedWord == checkWord):
            return (True, visited)
        print("Formed word: " + formedWord)
#         print(letterIndex)
        if(len(checkWord) > 1):
            letter = checkWord[letterIndex]
#             print("Checking for letter..." + letter)
            if(c<len(board[0])-1 and board[r][c+1] == letter):
#                 print("Next column entered...")
                c = c+1
                if((r, c) not in visited):
                    formedWord = formedWord + letter
                    visited.add((r, c))
                    (foundWord, visited) = self.checkLetter(board, r, c, checkWord, formedWord, letterIndex+1, visited)
                    if(foundWord == True):
                        return (True, visited)
                    else:
                        print("Continued after false..." + letter + "({row}, {col})".format(row = r, col = c))
            if(r<len(board)-1 and board[r+1][c] == letter):
                r = r+1
                if((r, c) not in visited):
                    formedWord = formedWord + letter
                    visited.add((r, c))
                    (foundWord, visited) = self.checkLetter(board, r, c, checkWord, formedWord, letterIndex+1, visited)
                    if(foundWord == True):
                        return (True, visited)
                    else:
                        print("Continued after false..." + letter + "({row}, {col})".format(row = r, col = c))
            if(board[r-1][c] == letter and r>0):
                r = r-1
                if((r, c) not in visited):
                    formedWord = formedWord + letter
                    visited.add((r, c))
                    (foundWord, visited) = self.checkLetter(board, r, c, checkWord, formedWord, letterIndex+1, visited)
                    if(foundWord == True):
                        return (True, visited)
                    else:
                        print("Continued after false..." + letter + "({row}, {col})".format(row = r, col = c))
            if(board[r][c-1] == letter and c>0):
                c = c-1
                if((r, c) not in visited):
                    formedWord = formedWord + letter
                    visited.add((r, c))
                    (foundWord, visited) = self.checkLetter(board, r, c, checkWord, formedWord, letterIndex+1, visited)
                    if(foundWord == True):
                        return (True, visited)
                    else:
                        print("Continued after false..." + letter)
            if((r, c) not in visited):
                print("First letter recursion started again...")
                formedWord = formedWord + letter
                if(self.checkLetter(board, r, c, checkWord, formedWord, letterIndex, visited)[0] == True):
                    return (True, visited)
#                 else:
#                     print("Continued after false..." + letter + "({row}, {col})".format(row = r, col = c))
            return (False, visited)

    def printBoard(self, board):
        for r in range(len(board)):
            print("\n")
            for c in range(len(board[0])):
                print(board[r][c], end="\t")
        print("\n")
```

```python
class Solution:
    def findWords(self, board, words):
        
        WORD_KEY = '$'
        
        # Building the trie
        trie = {}
        
        for word in words:
            node = trie
            for letter in word:
                node = node.setdefault(letter, {})
            node[WORD_KEY] = word
            
        rowNum = len(board)
        colNum = len(board[0])
        
        matched_words = []
        
        # Build the backtracking functionality
        def backtracking(row, col, parent):
            
#             print("Row : {row} ; Col : {col}".format(row=row, col=col))
            
            letter = board[row][col]
            currNode = parent[letter]
            
#             print("\nBoard letter: " + letter)
#             print("Current Node: " + str(currNode.keys()))
            
            # Mark the current cell as visited
            board[row][col] = '#'
            
            match_word = currNode.pop(WORD_KEY, False)
            
            if(match_word):
                matched_words.append(match_word)
                
            for (rowOffset, colOffset) in [(-1, 0), (0, -1), (1, 0), (0, 1)]:
                
                currRow = row + rowOffset
                currCol = col + colOffset
                
                if(currRow < 0 or currRow >= rowNum or currCol < 0 or currCol >= colNum):
                    continue
                if(board[currRow][currCol] not in currNode):
                    continue
#                 print("\nBoard letter: " + board_letter)
#                 print("Current Node: " + str(currNode.keys()))
                backtracking(currRow, currCol, currNode)
                
            board[row][col] = letter
            
#             if not currNode:
#                 parent.pop(letter)
                
#             print("Matched words: " + str(matched_words))
                
        for row in range(rowNum):
            for col in range(colNum):
                if(board[row][col] in trie):
                    backtracking(row, col, trie)
                
        return matched_words
    
    def printBoard(self, board):
        for r in range(len(board)):
            print("\n")
            for c in range(len(board[0])):
                print(board[r][c], end="\t")
        print("\n")
```
